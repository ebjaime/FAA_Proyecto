```python
def dataframe_medias_varianzas_por_id(sensor_d=sensor_d):
    dict_aux = {}
    for induction in sensor_d.id.unique():
        aux = []
        aux.append(induction)
        for attr in sensor_d.columns[2:12]:
            aux.append(sensor_d[sensor_d.id==induction][attr].mean())
            aux.append(sensor_d[sensor_d.id==induction][attr].var())
        dict_aux[induction] = aux
    return pd.DataFrame.from_dict(dict_aux, orient='index',columns=["id","R1_mean","R1_var","R2_mean",
                                                                "R2_var","R3_mean","R3_var","R4_mean",
                                                                "R4_var","R5_mean","R5_var","R6_mean","R6_var",
                                                                "R7_mean","R7_var","R8_mean","R8_var",
                                                                "Temp._mean","Temp._var",
                                                                "Humidity_mean", "Humidity_var"])
```
